```python
import numpy as np
class tictactoe:
    def __init__(self,player1,player2):
        self.name1=player1
        self.name2=player2
        self.board=[]
        self.board1()
     
    def board1(self):
        for i in range (0,3):
            row=[]
            for j in range (0,3):
                  row.append('_')
            self.board.append(row)
        print(np.array([self.board]))
                  
    def update_board(self,p_r,p_c,player): #p_r=row,p_c=column number
        if player==1:
            if self.board[p_r][p_c]=='_':  #To make sure player isn't overwriting already used spot
                self.board[p_r][p_c]='x'
            else:
                print('\n Try a different spot')
                self.input_row_column(1) #Go back and take the same player's inputs once again
                
                
        else:
            if self.board[p_r][p_c]=='_':
                self.board[p_r][p_c]='o'
            
            else:
                print('\n Try a different spot')
                self.input_row_column(2)
       
        print('\n',np.array([self.board]))
    
    def check_win(self):
        j=0
        for i in range(3):#horizontal    
            if self.board[i][j]==self.board[i][j+1] and self.board[i][j+1]==self.board[i][j+2] and self.board[i][j]!='_':
                return True
                
       
        for i in range(3):#vertical
            if self.board[j][i]==self.board[j+1][i] and self.board[j+1][i]==self.board[j+2][i] and self.board[j][i]!='_':
                return True
                
       
        #Diagonals
        if self.board[0][0]==self.board[1][1] and self.board[1][1]==self.board[2][2] and self.board[0][0]!='_':
            return True
            
        elif self.board[0][2]==self.board[1][1] and self.board[1][1]==self.board[2][0] and self.board[1][1]!='_':
            return True
           
       
    def board_full(self):#To check if board is filled
        for i in range(3):
            for j in range (3):
                if self.board[i][j]=='_':
                    return False
        else:
            return True
   
    def input_row_column(self,switch):# To get coordinate value from player and convert it to row and column no.
        co_ordinates=input("\nEnter the co-ordinate:")
        r=int(co_ordinates[0])
        c=int(co_ordinates[1])
        self.update_board(r,c,switch)
        
   
    def start(self):
        s=1 #Variable to help alternate between the players
        while(self.board_full()==False):
            if(s==1):
                print('\n',self.name1,"it's your turn:")
                self.input_row_column(s)
               
                if self.check_win()==True:
                    print(self.name1,'is Winner')
                    break
                if(self.board_full()==True):
                    print("GameOver: It's a Draw")
                    break
                s+=1
                
            if(s==2):
                print('\n',self.name2,"it's your turn:")
                self.input_row_column(s)
                if self.check_win()==True:
                    print(self.name2,'is Winner')
                    break
                if(self.board_full()==True):
                    print("GameOver: It's a Draw")
                    break
                s-=1
                      
        

p1=input('Enter name of player 1:')
p2=input('Enter name of player 2:')
print("\nCo-ordinates of the Tic Tac Toe grid:")
print("--------")
print("00 01 02\n10 11 12\n20 21 22\n")
t=tictactoe(p1,p2)
t.start()





        
        
        
        
```
